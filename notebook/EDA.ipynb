```python
def converRange(x):
    temp = x.split('-')  # Step 1: Split string by '-'
    if len(temp) == 2:    # Step 2: Agar 2 parts mile (range hai 1200-1500 )
        return (float(temp[0]) + float(temp[1])) / 2  # Average nikaal lo
    try:
        return float(x)   # Step 3: Agar single number hai, float me convert karo
    except:
        return None       # Step 4: Agar kuch bhi kaam na kare, return None
```

```python
df["total_sqft"] = df["total_sqft"].apply(converRange)
```

```python
def remove_outliers_sqft(df):
    cleaned_data = []

    # Go through each location one by one
    for location, group in df.groupby("location"):
        mean_price = group["price_per_square"].mean()
        std_price = group["price_per_square"].std()

        # Keep only rows within [mean - std, mean + std]
        filtered = group[
            (group["price_per_square"] > mean_price - std_price) & (group["price_per_square"] <= mean_price + std_price)
        ]

        # Add filtered rows to the list
        cleaned_data.append(filtered)

    # Combine all filtered groups into one DataFrame
    return pd.concat(cleaned_data, ignore_index=True)

```

```python
df=remove_outliers_sqft(df)
```

```python
def bhk_outlier_remover(df):
    # Step 1: Create an empty list to store indices of outlier rows
    outliers = []

    # Step 2: Loop through each unique location in the dataset
    for location in df["location"].unique():
        # Step 3: Get all rows for the current location
        loc_data = df[df["location"] == location]

        # Step 4: Loop through each unique BHK value in this location
        for bhk in loc_data["bhk"].unique():
            # Step 5: Get data for (BHK - 1) flats in the same location
            lower_bhk = loc_data[loc_data["bhk"] == bhk - 1]

            # Step 6: Get data for current BHK flats
            current_bhk = loc_data[loc_data["bhk"] == bhk]

            # Step 7: Only proceed if there are more than 5 (BHK - 1) flats
            if len(lower_bhk) > 5:
                # Step 8: Calculate average price per sqft for (BHK - 1)
                mean_lower = lower_bhk["price_per_square"].mean()

                # Step 9: Find current BHK flats priced lower than (BHK - 1) average
                bad = current_bhk[current_bhk["price_per_square"] < mean_lower]

                # Step 10: Add their indices to the outliers list
                outliers.extend(bad.index)

    # Step 11: Drop all outlier rows from the original DataFrame
    return df.drop(outliers)
```

```python
df=bhk_outlier_remover(df)
```

- Location-wise check karta hai
- Higher BHK flats ko remove karta hai agar unka price per sqft lower BHK se kam ho
- Sirf tab compare karta hai jab lower BHK ka data reliable ho (count > 5)
